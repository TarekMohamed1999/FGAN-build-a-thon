# 2. Parsing tables of use cases:

```python
"""
Parsing Tables of use cases
"""
for t in document.tables[:]:
    if t.cell(0,0).text == "Use case id":
        print(f"Use Case: {t.cell(0,1).text}")
        use_case = {}
        for r in t.rows:
            use_case[r.cells[0].text] = r.cells[1].text
        usecase.append(use_case)
        print("Table is Parsed well")
```

```python
usecase_df = pd.DataFrame()
```

```python
"""
Converting parsed data into a datframe
"""
for u in usecase:
    new_u = {}
    for key in u:
        new_u[key] = [u[key]]
    new_u = pd.DataFrame(new_u)
    usecase_df = pd.concat([usecase_df, new_u], ignore_index = True)
```

# 4. Parsing requirments:

```python
"""
Parsing requirements to be a dataframe
"""

reqs = {}
for p in document.paragraphs:
    if "REQ" in p.text:
        splits = p.text.split("-")
        req_number, req_content = splits[3].split(":")
        if len(splits) > 4:
            req_content += "-" + "-".join(splits[4:])
        if splits[1] in reqs:
            reqs[splits[1]][req_number] = req_content
        else:
            reqs[splits[1]] = {}
            reqs[splits[1]][req_number] = req_content
```

```python
len(reqs)
```

```python
# Converting into Dataframe
reqs_df = pd.DataFrame(reqs)
```

```python
reqs_df = reqs_df.T
```

```python
reqs_df
```

```python
def get_id(use_case_id):
    """
    This function to get the correct use case ID
    """
    splits = use_case_id.split("-")
    return f"UC{splits[-1]}"
```

```python
usecase_df["id"] = usecase_df["Use case id"].apply(get_id)
```

```python
# usecase_df
```

```python
usecase_df.loc[usecase_df["id"] == "UC9", "id"] = "UC009" 
```

```python
def edit_id(uid):
    """
    This function to fix the use case ID to be consistent in format
    """
    uid = uid.strip()
    if len(uid) != 5:
        splits = uid.split("C")
        if len(splits[1]) == 1:
            splits[1] = "00"+splits[1]
        elif len(splits[1]) == 2:
            splits[1] = "0"+splits[1]
        u_id = "UC" + splits[1]
        return u_id
    else:
        return uid
```

```python
usecase_df["id"] = usecase_df["id"].apply(edit_id)
```

```python
usecase_df.drop(columns = ["new_id"], inplace = True)
```

# 5. Joining Requirements and use case data:

```python
reqs_df = reqs_df.reset_index().rename(columns = {"index":"id"})
```

```python
def id_extended(id_):
    """
    Final shape of the use case ID to join the dataframes on
    """
    if len(id_) < 5:
        splits = id_.split("C")
        return f"UC0{splits[1]}"
    else:
        return id_
```

```python
reqs_df["id"] = reqs_df["id"].apply(id_extended)
```

# 6. Getting requirements importance:

```python
def get_importance(req):
    """
    This function is to get the importance of requirements.
    """
    if type(req) == str:
        splits = req.split(" ")
        if splits[3].lower() in ["critical", "expected"]:
            return splits[3]
        elif splits[4].lower() in ["added"]:
            return "added value"
    else:
        return np.nan
```

```python
usecases_df2 = usecase_df1.copy()
```

```python
"""
Renaming the importance columns
"""
for index, row in usecases_df2.iterrows():
    for col in row.index:
        if "0" in col:
            req_importance = col + "_importance"
            usecases_df2.loc[index , req_importance] = get_importance(usecases_df2.loc[index , col])
        
```
