# 1. Figure Parsing:

```python
def figure_parse(path):
    """
    Parsing XML figures.
    input: Path that contains files
    output: no output, but editing the required dicts.
    """
    tree = ET.parse(path)
    root = tree.getroot()
    uid = path.partition(".xml")[0].split("/")[-1]
    # Extracting information
    functions = []
    realtions = [] 
    components = []
    for mcell in root.findall('diagram/mxGraphModel/root/object'):
        if (mcell.attrib["class"] == "component") | (mcell.attrib["class"] == "actor"):
            components.append({mcell.attrib["id"]: mcell.attrib["value"]})
        elif mcell.attrib["class"] == "function":
            functions.append({mcell.attrib["value"]: mcell.attrib["parent"]})
        else:
            cell = mcell.find("mxCell")
            source = cell.attrib["source"]
            target = cell.attrib["target"]
            message = mcell.attrib["value"]
            class_ = mcell.attrib["class"]
            print(f"source: {source} <==={message}===> ", f"target: {target} , ({class_})")
            realtions.append({"source":source,"target":target,
                                  "message":message,"class":class_})
    realtions_df1 = pd.DataFrame(realtions)
#     realtions_df1["source"] = realtions_df1.apply(lambda x: get_component(realtions_df1["source"],
#                                                                            components), axis = 1)
#     realtions_df1["target"] = realtions_df1.apply(lambda x: get_component(realtions_df1["target"],
#                                                                            components), axis = 1)
    realtions_df1["id"] = uid
    global realtions_df
    realtions_df = pd.concat([realtions_df, realtions_df1], ignore_index = True)
    global compoenents_dict
    compoenents_dict[uid] = components
    global functions_dict
    functions_dict[uid] = functions
```

# 2. Get Component names:

```python
def get_component(id_):
    """
    To get the name of a specific Component
    """
    for com in get_components.comps:
        if id_ in com:
            return com[id_]
def get_components(rels):
    """
    getting components.
    """
    global compoenents_dict, new_df
    us = rels.groupby("id")
    ids = list(us.groups.keys())
    for g in ids:
        u = us.get_group(g)
        get_components.comps = compoenents_dict[g]
        u["source"] = u["source"].apply(get_component)
        u["target"] = u["target"].apply(get_component)
        new_df = pd.concat([new_df, u], ignore_index = True)
```

# 3. Getting Parents for functions:

```python
def get_component(id_, uid):
    
    """
    This function gets the component name from its ID
    input: component Id (string)
            use case id (string)
    output: component name (string)
    """
    
    global compoenents_dict
    for com in compoenents_dict[uid]:
        if id_ in com:
            return com[id_]
```
